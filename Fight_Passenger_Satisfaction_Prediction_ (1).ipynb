```python
def cap_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)

cap_outliers(df, 'Flight Distance')
cap_outliers(df, 'Departure Delay in Minutes')
cap_outliers(df, 'Arrival Delay in Minutes')
```

select x(independent features) and y(target)

```python
x=df.drop(columns=['satisfaction'])
y=df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
le_customer=LabelEncoder()
le_travel=LabelEncoder()
le_class=LabelEncoder()
df['Gender']=le_gender.fit_transform(df['Gender'])
df['Customer Type']=le_customer.fit_transform(df['Customer Type'])
df['Type of Travel']=le_travel.fit_transform(df['Type of Travel'])
df['Class']=le_class.fit_transform(df['Class'])
```

**Synthetic Minority Oversampling Technique(SMOTE)**

**Model training using Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression(max_iter=1000)
```

```python
LR.fit(x_train,y_train)
```

```python
prediction=LR.predict(x_test)
```

To evaluate classification model performance

```python
from sklearn.metrics import *
accuracy_score(y_test,prediction)
```

```python
classification_report(y_test,prediction)
```

**Model training using Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
```

```python
Predict=DT.predict(x_test)
```

```python
accuracy_score(y_test,Predict)
```

```python
classification_report(y_test,Predict)
```

Model training using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier()
```

```python
RF.fit(x_train,y_train)
```

```python
predict=RF.predict(x_test)
```

```python
accuracy_score(y_test,predict)
```

```python
classification_report(y_test,predict)
```

Among the three Machine Learning Algorithms i.e Logistic Regression,Decision tree,Random forest classifier the accuracy of random forest classifier has best score.Hence it is used in ML model web app
